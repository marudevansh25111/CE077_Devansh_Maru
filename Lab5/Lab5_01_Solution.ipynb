# **Task 1: Try the algo on Same Weather dataset - OneHotEncoding of features: and Train test Division 75%-25%**

```python
import numpy as np
import pandas as pd
import sklearn

from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
```

```python
#Predictor variables
Outlook = ['Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Sunny', 'Overcast',
'Rainy', 'Rainy', 'Sunny', 'Rainy','Overcast', 'Overcast', 'Sunny']

Temperature = ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool',
'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild']

Humidity = ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal',
'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High']

Wind = ['False', 'True', 'False', 'False', 'False', 'True', 'True',
'False', 'False', 'False', 'True', 'True', 'False', 'True']

#Class Label:
Play = ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No',
'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
```

```python
#creating OneHotEncoder
oe = preprocessing.OneHotEncoder();

# Converting string labels into numbers.
Outlook_encoded = oe.fit_transform(np.array(Outlook).reshape(-1,1)).toarray()

Temperature_encoded = oe.fit_transform(np.array(Temperature).reshape(-1,1)).toarray()

Humidity_encoded = oe.fit_transform(np.array(Humidity).reshape(-1,1)).toarray()

Wind_encoded = oe.fit_transform(np.array(Wind).reshape(-1,1)).toarray()

Play_encoded = oe.fit_transform(np.array(Play).reshape(-1,1)).toarray()

df1 = pd.DataFrame(Outlook_encoded, columns=["Outlook_"+str(int(i)) for i in range(len(Outlook_encoded[0]))])
df2 = pd.DataFrame(Temperature_encoded, columns=["Temp_"+str(int(i)) for i in range(len(Temperature_encoded[0]))])
df3 = pd.DataFrame(Humidity_encoded, columns=["Wind_"+str(int(i)) for i in range(len(Humidity_encoded[0]))])
df4 = pd.DataFrame(Wind_encoded, columns=["Humidity_"+str(int(i)) for i in range(len(Wind_encoded[0]))])



df = pd.concat([df1,df2,df3,df4], axis=1)
df
```

```python
dataset = df.iloc[:,:].values
print(dataset)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset, Play_encoded, test_size=0.25)

print("X_train = ", x_train)
print("X_test = ", x_test)
print("Y_train = ", y_train)
print("Y_train = ", y_test)
```

```python
classifier =  DecisionTreeClassifier(criterion='entropy', random_state=77, max_leaf_nodes=77)
clf_entropy = classifier.fit(x_train, y_train)

from sklearn import tree

text_representation = tree.export_text(classifier)
print(text_representation)
```

```python
y_pred = classifier.predict(x_test)

y_pred=y_pred.flatten()
y_test=y_test.flatten()

print("y_predict = ", y_pred)
print("y_test = ", y_test)

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score

accuracy = accuracy_score(y_test, y_pred, normalize=True)
print("\n\naccuracy = ", accuracy)

pricison, recall, f1score, support = score(y_test, y_pred, average='micro')

print("Pricison = ", pricison)
print("Recall = ", recall)
print("F1Score = ", f1score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz

export_graphviz(clf_entropy, out_file='tree_entropy.dot', 
                feature_names=['Outlook_Rainy', 'Outlook_Sunny', 'Outlook_Overcast', 'Temperature_Hot', 'Temperature_Mild', 'Temperature_Cool', 'Humidity_High', 'Humidity_Normal', 'Wind_True', 'Wind_False'], 
                class_names=['play_no', 'play_yes'],
                filled=True)


#convert to png
from subprocess import call

call(['dot', '-Tpng', 'tree_entropy.dot', '-o', 'tree_entropy.png', '-Gdpi=600'])

import matplotlib.pyplot as plt
plt.figure(figsize = (14,18))
plt.imshow(plt.imread('tree_entropy.png'))
plt.axis('off')
plt.show()
```
