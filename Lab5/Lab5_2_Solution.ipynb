```python
#Import scikit-learn dataset library
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import numpy as np

#Load dataset
digits = datasets.load_digits()

digits
```

```python
print("Features of iris dataset :- \n", digits.feature_names)

print("\nLabel types of iris dataset :-\n", digits.target_names)

print("\nIris dataset shape :- ", digits.data.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.8)

print("X_train = ", x_train)
print("\nX_test = ", x_test)
print("\nY_train = ", y_train)
print("\nY_train = ", y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy', random_state=96, max_leaf_nodes=96)
clf_entropy = classifier.fit(x_train, y_train)

from sklearn import tree

text_representation = tree.export_text(classifier)
print(text_representation)
```

```python
y_pred = classifier.predict(x_test)

print("y_predict = ", y_pred)
print("y_test = ", y_test)

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score

accuracy = accuracy_score(y_test, y_pred, normalize=True)
print("\n\naccuracy = ", accuracy)

pricison, recall, f1score, support = score(y_test, y_pred, average='micro')

print("Pricison = ", pricison)
print("Recall = ", recall)
print("F1Score = ", f1score)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz

export_graphviz(clf_entropy, out_file='digits_tree.dot', 
                feature_names=list(digits.feature_names), 
                class_names=list(str(digits.target_names)),
                filled=True)


#convert to png
from subprocess import call

call(['dot', '-Tpng', 'digits_tree.dot', '-o', 'digits_tree.png', '-Gdpi=600'])

import matplotlib.pyplot as plt
plt.figure(figsize = (14,18))
plt.imshow(plt.imread('digits_tree.png'))
plt.axis('off')
plt.show()
```
