3) Get the Transpose of the Matrix that you created. Create a square matrix and find its determinant.

```python
mat3=np.array([[1,2],[3,4]])
print("Transpose matrix:")
print(mat3.T)
print("square array:")
sqkmatrix = mat3 * mat3
print(sqkmatrix)
print("Value of determinent is:")
det=np.linalg.det(sqkmatrix)
print(int(det))
```
